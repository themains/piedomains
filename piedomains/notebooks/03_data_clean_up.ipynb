# Process .html files

```python
from bs4 import BeautifulSoup
from bs4.element import Comment
```

```python
def tag_visible(element):
    if element.parent.name in ['style', 'script']:
        return False
    if isinstance(element, Comment):
        return False
    return True

def text_from_html(file):
    with open(file, 'r', encoding="utf8", errors='ignore') as f:
        contents = f.read()
        soup = BeautifulSoup(contents, 'lxml')
        texts = soup.findAll(text=True)
        visible_texts = filter(tag_visible, texts)
        result = u" ".join(t.strip() for t in visible_texts)
        return ' '.join(result.split())
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer 

import nltk
import re
import string

nltk.download('stopwords')
nltk.download('words')
words = set(nltk.corpus.words.words())
stop_words = set(stopwords.words('english'))
most_common_words = ['home', 'contact', 'us', 'new', 'news', 'site', 'privacy', 'search', 'help', 'copyright', 'free', 'service', 'en', 'get', 'one', 'find', 'menu', 'account', 'next']

def data_cleanup(s):
  # remove numbers
  s = re.sub(r'\d+', '', s)
  # remove duplicates
  tokens = list(set(s.split()))
  tokens.sort()
  # remove punctuation from each token
  table = str.maketrans('', '', string.punctuation)
  tokens = [w.translate(table) for w in tokens]
  # remove non english words
  tokens = [w.lower() for w in tokens if w.lower() in words]
  # remove non alpha
  tokens = [w for w in tokens if w.isalpha()]
  # remove non ascii
  tokens = [w for w in tokens if w.isascii()]
  # filter out stop words
  tokens = [w for w in tokens if not w in stop_words]
  # filter out short tokens
  tokens = [word for word in tokens if len(word) > 1]
  # remove most common words
  tokens = [w for w in tokens if not w in most_common_words]
  tokens = list(set(tokens))
  return " ".join(w for w in tokens)
```

```python
# this is to track what domains are loaded 
loaded = []
try:
  with open('/content/drive/MyDrive/Colab/pydomains/data/domain_text_v3.csv', 'r') as f:
    for line in f:
      loaded.append(line.split(',')[0])
except:
  pass
```

```python
import os
import csv

directory = '.'

count = 0
with open('/content/drive/MyDrive/Colab/pydomains/data/domain_text_v3.csv', 'a+') as f:
    writer = csv.writer(f)
    #  if already some are processed, skip this
    if len(loaded) == 0:
      writer.writerow(['domain','text', 'cat_name'])
    
    rows = []
    for root, dirs, files in os.walk(directory):
        for filename in files:
            if '.html' in filename:
                if count % 10000 == 0:
                  writer.writerows(rows)
                  rows = []
                  print(f"{count} domains parsed")
                # get just the domain name
                domain = filename.strip(".html").rsplit(".",1)[0]
                # get the category
                cat_name = ''
                try:
                  cat_name = cat_dict[domain]
                except:
                  pass
                #  if already some are processed, skip this
                if domain not in loaded:
                  # get the text in that domain
                  result = data_cleanup(text_from_html(filename))
                  rows.append([domain,result,cat_name])
                count = count + 1
```

# Vocab

```python
from collections import Counter
  
cat_vocab = {}

def build_cat_vocab(cat, text):
  if cat not in cat_vocab:
    cat_vocab[cat] = Counter()
  cat_vocab[cat].update(text.split())
  

for index, row in domain_text.iterrows():
  build_cat_vocab(row['cat_name'], row['text'])

#new_df['text'].map(lambda x: vocab.update(x.split()))
```
