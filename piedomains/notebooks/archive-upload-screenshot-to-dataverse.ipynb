```python
from pyDataverse.api import Api
from pyDataverse.models import Dataverse
api = Api('https://dataverse.harvard.edu/', token)
api.status
```

```python
resp = api.get_dataset(doi)
resp.json()
```

```python
from glob import glob
import os
```

```python
files = {}
for fn in glob('../scripts/png/*.jpg'):
    n = int(fn.split('/')[-1].replace('.jpg', ''))
    files[n] = fn
```

```python
len(files)
```

```python
max(files)
```

```python
%%time
import tarfile

COUNT = 10000

for i in range(1, 25):
    beg = (i - 1) * COUNT
    end =  i * COUNT
    tarfn = 'screenshot-%02d.tar.gz' % i
    print(i, beg, end, tarfn)
    with tarfile.open(tarfn, "w:gz") as t:
        for j in range(beg, end):
            if j in files:
                t.add(files[j], arcname='screenshot/' + os.path.basename(files[j]))
    resp = api.upload_file(doi, tarfn)
    print(resp)
    os.unlink(tarfn)
    #break
```
