```python
cat_df = cat_df.dropna()
```

```python
cat_df = cat_df.groupby('private_domain')['category'].apply(','.join).reset_index()
```

```python
cat_df.count()
```

```python
final = cat_df.merge(df, left_on='private_domain', right_on = 'domain', how="inner")
```

```python
final.head()
```

```python
final = final.drop(['Unnamed: 0.1','Unnamed: 0'], axis=1)
```

```python
actual_cats = list(set(','.join(list(final['category'].unique())).split(',')))
```

```python
def match_percent(actual, predicted):
  matched = final[final['category'].str.contains(actual)][final[final['category'].str.contains(actual)]['label'] == predicted]['label'].count()
  total  = final[final['category'].str.contains(actual)]['label'].count()
  return matched/total
```

```python
match_percent("Adult", "porn")
```

```python
match_percent("Shopping", "shopping")
```
